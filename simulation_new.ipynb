# Simulation

Perform basic analyses on a simulation

```python
import matplotlib.pyplot as plt
import numpy as np

from bluepy.v2 import Simulation
```

```python
sim = Simulation('circuit/simulation/20180907/BlueConfig')
```

```python
gids = sim.circuit.cells.ids('SP_PC', sample=100)
```

```python
sim.plot.raster(gids, t_start=0);
```

```python
sim.plot.firing_rate_histogram(gids, t_start=0);
```

```python
sim.plot.isi(gids, t_start=0);
```

```python
sim.plot.voltage_collage('soma', group=gids[:5], t_start=0);
```
